# Data Analysis and Visualization

```python
cancelled_percentage = df['is_canceled'].value_counts(normalize = True)
print(cancelled_percentage)

plt.figure(figsize = (6,5))
plt.title('Reservation status add up')
plt.bar(['Not Cancelled', 'Cancelled'],df['is_canceled'].value_counts(),edgecolor = 'k', width = 0.7)
plt.show()
```

Above graph shows that the rate of cancellation of the reservations are more than half of the rate of non cancellation

```python
plt.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
legend_lables,_ =ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not cancelled','cancelled'])
plt.show()
```

Hence the ratio of cancelling and non cancelling a reservation in city hotels is much more than resort hotels. 
High prices in resort hotels can be on of the major resons in their cancellation. 
In city hotels the reasons can be with their poor maintenance of rooms and other facilities.

```python
resort_hotels = df[df['hotel'] == 'Resort Hotel']
resort_hotels['is_canceled'].value_counts(normalize = True)
```

```python
city_hotels = df[df['hotel'] == 'City Hotel']
city_hotels['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotels = resort_hotels.groupby('reservation_status_date')[['adr']].mean()
city_hotels = city_hotels.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,10))
plt.title('Average Daily Rate in Resort and City Hotels', fontsize = 30)
plt.plot(resort_hotels.index, resort_hotels['adr'], label = 'Resort Hotel')
plt.plot(city_hotels.index,city_hotels['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

Thus, graph depicts, the Average daily rate of city hotels are lesser as compared to resort hotels.

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month',hue = 'is_canceled', data= df, palette = 'bright')
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1))
plt.title('Reservation status per month', size = 20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not cancelled','cancelled'])
plt.show()
```

so, we can see that, most of the cancellations are happening in the month of January, along with least no. of reversations.
Whereas, in the month of August, most of the reservations took place, along with least no. of cancellations.
one of the reasons might be with their adrs on these perticular months:-

```python
plt.figure(figsize = (15,8))
plt.title('ADR per month', fontsize = 30)
sns.barplot('month','adr',data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

Hence it shows that, higher adr leads to higher cancellations, whereas lower adrs lead to low the cancellations. Which means that they are directly propotional to each other.

```python
cancelled_data = df[df['is_canceled'] == 1]
top_10_countries = cancelled_data['country'].value_counts()[:10]
print(top_10_countries)
plt.figure(figsize = (8,8))
plt.title('Top 10 countries with reservation cancelled')
plt.pie(top_10_countries,autopct = '%.2f', labels = top_10_countries.index)
plt.show()
```

Above graph represents that hotels in Portugal faces the most cancellations of reservations. 
Hence to reduce this issue, the suggestions can be to look into their room facilities and their prices,
should give promotion discounts or run campaigns, advertisements and marketing stretagies to deal with it.

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label = 'cancelled')
plt.legend()
plt.show()
```

```python
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize = (20,6))
plt.title('Average Daily Rate', fontsize = 30)
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label = 'cancelled')
plt.legend(fontsize = 20)
plt.show()
```

Again we can see that rate of cancellations are more by the increasing Avg daily rate.
